```

from fastai.vision.all import *
from fastai.vision.widgets import *
def parent_label_multi(o):
    return [Path(o).parent.name]

path = Path()
learn_inf = load_learner(path/'invasive_plants_II/fmodel.pkl', cpu=True)
btn_upload = widgets.FileUpload()
btn_run = widgets.Button(description='Classify')
out_pl = widgets.Output()


def on_click_classify(change):
    result = ""
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()

    labels, prediction, probability = learn_inf.predict(img)
    for idx,label in enumerate(learn_inf.dls.vocab):
      result = result + f"[category: {label}, prediction: {prediction[idx]}, probablity: {probability[idx]:.2f} ]"
    with out_pl: display(result)

btn_upload.observe(on_click_classify, names=['data'])
display(VBox([widgets.Label('Select'), btn_upload, out_pl, out_text]))
```
